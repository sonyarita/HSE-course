2. –í—ã–ø–æ–ª–Ω–∏—Ç–µ —Ç–æ–∫–µ–Ω–∏–∑–∞—Ü–∏—é –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ–≥–æ —Ç–µ–∫—Å—Ç–∞ –Ω–∞ —Å—Ç—Ä–æ–∫–∏ –ø—Ä–∏ –ø–æ–º–æ—â–∏ split('\n'). –û—Ç–æ–±—Ä–∞–∑–∏—Ç–µ —Å–∫–æ–ª—å–∫–æ –∏—Ö.

```python
lines_text = text.split('\n') # –¥–µ–ª—é —Ç–µ–∫—Å—Ç –Ω–∞ —Å—Ç—Ä–æ–∫–∏
print(lines_text) # —Å–º–æ—Ç—Ä—é –∫–∞–∫–∏–µ —Å—Ç—Ä–æ–∫–∏ –ø–æ–ª—É—á–∏–ª—Å—å
print(len(lines_text)) # —Å—á–∏—Ç–∞—é –∫–æ–ª–∏—á–µ—Å–≤—Ç–æ —Å—Ç—Ä–æ–∫
```

3. –ù–∞–ø–∏—à–∏—Ç–µ —Ü–∏–∫–ª, –∫–æ—Ç–æ—Ä—ã–π –ø—Ä–æ–π–¥–µ—Ç –ø–æ –≤—Å–µ–º —Ç–æ–∫–µ–Ω–∞–º —Ç–µ–∫—Å—Ç–∞ –∏ —Å–æ—Ö—Ä–∞–Ω–∏—Ç –≤ –Ω–æ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π —Ç–µ, –∫–æ—Ç–æ—Ä—ã–µ –Ω–∞—á–∏–Ω–∞—é—Ç—Å—è —Å –∑–∞–≥–ª–∞–≤–Ω–æ–π –±—É–∫–≤—ã.

```python
words = nltk.word_tokenize(text) # —Ç–æ–∫–µ–Ω–∏–∑–∏—Ä—É—é —Ç–µ–∫—Å—Ç
capitalized_words = [] # —Å–æ–∑–¥–∞—é –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫, –≤ –∫–æ—Ç–æ—Ä—ã–π –ø–æ—Ç–æ–º –¥–æ–±–∞–≤–ª—é —Å–ª–æ–≤–∞ —Å –∑–∞–≥–ª–∞–≤–Ω–æ–π –±—É–∫–≤—ã
for word in words: # –ø–µ—Ä–µ–±–∏—Ä–∞—é —Ü–∏–∫–ª–æ–º —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π words 
    if word[0].isupper(): # –µ—Å–ª–∏ –ø–µ—Ä–≤—ã–π —ç–ª–µ–º–µ–Ω—Ç (–ø–æ–¥ –Ω—É–ª–µ–≤—ã–º –∏–Ω–¥–µ–∫—Å–æ–º) —Ç–æ–∫–µ–Ω–∞ —è–≤–ª—è–µ—Ç—Å—è –∑–∞–≥–ª–∞–≤–Ω–æ–π –±—É–∫–≤–æ–π, —Ç–æ
        capitalized_words.append(word) # –¥–æ–±–∞–≤–ª—è—é –≤ —Å–ø–∏—Å–æ–∫ —ç—Ç–æ—Ç —Ç–æ–∫–µ–Ω
print(len(capitalized_words)) # –ø–µ—á–∞—Ç–∞—é –¥–ª–∏–Ω—É —Å–ø–∏—Å–∫–∞
print(capitalized_words) # –∏ —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ —Å–ø–∏—Å–∫–∞
```

1. –ò–º–ø–æ—Ä—Ç–∏—Ä—É–π—Ç–µ —Å–ø–∏—Å–æ–∫ —Å—Ç–æ–ø-—Å–ª–æ–≤ –¥–ª—è —Ä—É—Å—Å–∫–æ–≥–æ —è–∑—ã–∫–∞ –∏–∑ NLTK.
2. –ù–∞–ø–∏—à–∏—Ç–µ —Ü–∏–∫–ª, –∫–æ—Ç–æ—Ä—ã–π –ø—Ä–æ–π–¥–µ—Ç –ø–æ –≤—Å–µ–º —Ç–æ–∫–µ–Ω–∞–º –∏ –≤–µ—Ä–Ω–µ—Ç –Ω–æ–≤—ã–π —Å–ø–∏—Å–æ–∫ —Ç–æ–∫–µ–Ω–æ–≤ –±–µ–∑ —Å—Ç–æ–ø-—Å–ª–æ–≤.
3. –°–∫–æ–ª—å–∫–æ —Å–ª–æ–≤ –æ—Å—Ç–∞–ª–æ—Å—å –≤ —Ç–µ–∫—Å—Ç–µ? –ê —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö?
4. *–ü—Ä–∏ –ø–æ–º–æ—â–∏ —Ä–µ–≥—É–ª—è—Ä–Ω–æ–≥–æ –≤—ã—Ä–∞–∂–µ–Ω–∏—è –Ω–∞–π–¥–∏—Ç–µ –≤—Å–µ —Å–ª–æ–≤–∞ —Å —Å–æ—á–µ—Ç–∞–Ω–∏–µ–º –±—É–∫–≤ '–±–æ–ª' (—Å –æ–ø—Ü–∏–æ–Ω–∞–ª—å–Ω–æ–π –æ) –≤–Ω—É—Ç—Ä–∏ —Å–ª–æ–≤–∞. –î–æ–±–∞–≤—å—Ç–µ –∏—Ö –∫ —Å—Ç–æ–ø-—Å–ª–æ–≤–∞–º –∏ —É–¥–∞–ª–∏—Ç–µ –∏–∑ —Ç–µ–∫—Å—Ç–∞. –°–∫–æ–ª—å–∫–æ –æ—Å—Ç–∞–ª–æ—Å—å —Å–ª–æ–≤ –≤ —Ç–µ–∫—Å—Ç–µ?

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
import re
stopwords_ru = stopwords.words('russian')
```

2. –ù–∞–ø–∏—à–∏—Ç–µ —Ü–∏–∫–ª, –∫–æ—Ç–æ—Ä—ã–π –ø—Ä–æ–π–¥–µ—Ç –ø–æ –≤—Å–µ–º —Ç–æ–∫–µ–Ω–∞–º –∏ –≤–µ—Ä–Ω–µ—Ç –Ω–æ–≤—ã–π —Å–ø–∏—Å–æ–∫ —Ç–æ–∫–µ–Ω–æ–≤ –±–µ–∑ —Å—Ç–æ–ø-—Å–ª–æ–≤.

```python
tokens = nltk.word_tokenize(normal_text)
no_stop_words = []
for token in tokens:
    if token not in stopwords_ru:
        no_stop_words.append(token)
print(no_stop_words)
print(len(no_stop_words))
```

3. –°–∫–æ–ª—å–∫–æ —Å–ª–æ–≤ –æ—Å—Ç–∞–ª–æ—Å—å –≤ —Ç–µ–∫—Å—Ç–µ? –ê —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö?

```python
print(len(no_stop_words)) # –≤—Å–µ–≥–æ —Å–ª–æ–≤
print(len(set(no_stop_words))) # —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö
```

4. *–ü—Ä–∏ –ø–æ–º–æ—â–∏ —Ä–µ–≥—É–ª—è—Ä–Ω–æ–≥–æ –≤—ã—Ä–∞–∂–µ–Ω–∏—è –Ω–∞–π–¥–∏—Ç–µ –≤—Å–µ —Å–ª–æ–≤–∞ —Å —Å–æ—á–µ—Ç–∞–Ω–∏–µ–º –±—É–∫–≤ '–±–æ–ª' (—Å –æ–ø—Ü–∏–æ–Ω–∞–ª—å–Ω–æ–π –æ) –≤–Ω—É—Ç—Ä–∏ —Å–ª–æ–≤–∞. –î–æ–±–∞–≤—å—Ç–µ –∏—Ö –∫ —Å—Ç–æ–ø-—Å–ª–æ–≤–∞–º –∏ —É–¥–∞–ª–∏—Ç–µ –∏–∑ —Ç–µ–∫—Å—Ç–∞. –°–∫–æ–ª—å–∫–æ –æ—Å—Ç–∞–ª–æ—Å—å —Å–ª–æ–≤ –≤ —Ç–µ–∫—Å—Ç–µ?

```python
bol_words =  re.findall(r'\w*–±[–æ]?–ª\w*', normal_text) # —Å –ø–æ–º–æ—â—å—é —Ä–µ–≥—É–ª—è—Ä–∫–∏ –∏—â—É —Å–ª–æ–≤–∞ —Å —Å–æ—á–µ—Ç–∞–Ω–∏–µ–º –±—É–∫–≤ '–±–æ–ª' (—Å –æ–ø—Ü–∏–æ–Ω–∞–ª—å–Ω–æ–π –æ)
print(bol_words)
```

```python
extended_stopwords = stopwords_ru + bol_words # –æ–±—ä–µ–¥–∏–Ω—è—é —Å—Ç–æ–ø-—Å–ª–æ–≤–∞ –∏ —Å–ª–æ–≤–∞ —Å –±–æ?–ª
clean_text = [] # —Å–æ–∑–¥–∞—é —Å–ø–∏—Å–æ–∫, –∫—É–¥–∞ –±—É–¥—É –≤—Å—ë –¥–æ–±–∞–≤–ª—è—Ç—å
for token in tokens: # –ø–µ—Ä–µ–±–∏—Ä–∞—é —Ü–∏–∫–ª–æ–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é
    if token not in extended_stopwords: # –µ—Å–ª–∏ —Ç–æ–∫–µ–Ω –Ω–µ –≤ —Ä–∞—Å—à–∏—Ä–µ–Ω–Ω–æ–º —Å–ø–∏—Å–∫–µ —Å—Ç–æ–ø-—Å–ª–æ–≤, —Ç–æ
        clean_text.append(token) # –¥–æ–±–∞–≤–ª—è—é –µ–≥–æ –≤ –Ω–æ–≤—ã–π —Å–ø–∏—Å–æ–∫
print(clean_text)
print(len(clean_text)) # –æ—Å—Ç–∞–≤—à–µ–µ—Å—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ª–æ–≤ –≤ —Ç–µ–∫—Å—Ç–µ
```
